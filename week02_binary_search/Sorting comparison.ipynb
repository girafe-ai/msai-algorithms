```python
import sys
sys.path.append('..')
sys.path.append('../../week01_sorting_algorithms/')
from qsort import qsort
from bubble import bubble_sort
from selection import selection_sort
from insertion import insertion_sort
from merge import merge_sort
from counting import counting_sort, counting_sort2
from random import randint
```

```python
for N in [1000, 10000, 100000, 1000000, 10000000, 100000000]:
    print(f'======= N={N} ======')
    X = [randint(0, 100) for i in range(N)]
    x = list(X)
    print('Built-in: ')
    %time x.sort()
    
    x = list(X)
    print('CountingSort: ')
    %time counting_sort(x)
    
    x = list(X)
    print('MergeSort: ')
    if N >= 100000000:
        print('TimeOut')
    else:
        %time merge_sort(x)
    
    x = list(X)
    print('QSort: ')
    if N >= 100000000:
        print('TimeOut')
    else:
        %time qsort(x)
    
    x = list(X)
    print('BubbleSort: ')
    if N >= 100000:
        print('TimeOut')
    else:
        %time bubble_sort(x)
```
