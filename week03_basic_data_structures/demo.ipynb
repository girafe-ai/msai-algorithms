```python
from time import time
from collections import deque
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt
```

## Vector (python list) adding to the end (append, push_back)

```python
ns = []
ts = []
for N in tqdm(list(range(1, 5 * 10**5, 1000))):
    tt = []
    for k in range(10):
        t = time()
        x = []
        for i in range(N):
            x.append(i)
        tt.append(time() - t)
    ns.append(N)
    ts.append(sum(tt) / 10)
plt.plot(ns, ts)
```

## Vector (python list) adding to the beginning

```python
ns = []
ts = []
for N in tqdm(list(range(1, 10**5, 1000))):
    t = time()
    x = []
    for i in range(N):
        x.insert(0, i)
    ns.append(N)
    ts.append(time() - t)
plt.plot(ns, medfilt(ts, 11))
```

## Doubly Linked List (collections.deque) adding to the beginning

```python
ns = []
ts = []
for N in tqdm(list(range(1, 10**6, 1000))):
    t = time()
    x = deque()
    for i in range(N):
        x.appendleft(i)
    ns.append(N)
    ts.append(time() - t)
plt.plot(ns, ts)
```
